```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
file_sal = 'salary_table.csv'
file_bat = 'team_batting.csv'
file_pitch = 'team_pitch.csv'
```

```python
salary_df = pd.read_csv(file_sal, encoding="utf-8")
batting_df = pd.read_csv(file_bat, encoding="utf-8")
pitch_df = pd.read_csv(file_pitch, encoding="utf-8")
```

```python
salary_df.head()
```

```python
salary_df = salary_df.drop(columns=['Unnamed: 0'])
```

```python
batting_df.head()
```

```python
batting_df = batting_df[['Tm', 'OPS','OBP']].drop(index=[30,31])
```

```python
batting_df = batting_df.rename(columns={'Tm': 'Teams'})
```

```python
batting_df
```

```python
pitch_df.head()
```

```python
pitch_df = pitch_df[['Tm', 'WHIP','ERA']].drop(index=[30,31])
```

```python
pitch_df = pitch_df.rename(columns={'Tm': 'Teams'})
```

```python
stats_df = pd.merge(salary_df, batting_df, how='inner', on='Teams')
```

```python
stats_df
```

```python
stats_df = pd.merge(stats_df, pitch_df, how='inner', on='Teams')
```

```python
stats_df
```

```python
stats_df.loc[len(stats_df)] = new_row
```
