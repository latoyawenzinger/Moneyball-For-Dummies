```python
from splinter import Browser
from bs4 import BeautifulSoup
from splinter import Browser
from selenium.webdriver.chrome.service import Service
import os
driver_path = r'C:\Users\lnorm\Downloads\chromedriver_win32 (2)\chromedriver.exe'
browser = Browser("chrome",service=Service(executable_path=driver_path))

import pandas as pd
```

```python
# open website
url = 'https://www.spotrac.com/mlb/payroll/2022/'
browser.visit(url)
```

```python
# create a BeutifulSoup object
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# extract all rows of data
tables = soup.find('table', class_='datatable rtable captracker tablesorter tablesorter-default tablesorter6f41d7d5da217')
rows = tables.find_all('tr')
# showing the first 3 rows of data to confirm data was extracted
print(rows[:3])
```

```python
# create an empty list
table_list = []

# loop through the scraped data to create a list of rows

for row in rows:
    td = row.find_all('td')
    row = [row.text.strip() for row in td if row.text.strip()]
    if row:
        table_list.append(row)
        
salary_table = pd.DataFrame(table_list, columns=['Rank', 'Team', 'Wins%', 
                                               'Roster', '26-Man Payroll', 'Injured Reserve', 'Retained', 'Buried', 'Suspended', '2022 Total Payroll'])
```

```python
salary_table
```

```python
clean_salary_table = salary_table.drop(columns=['Rank', 'Roster', '26-Man Payroll', 'Injured Reserve', 'Retained', 'Buried', 'Suspended'],
                  index=[14,15])
```

```python
clean_salary_table['Teams'] = [x.split('\n')[0] for x in clean_salary_table['Team']]
```
